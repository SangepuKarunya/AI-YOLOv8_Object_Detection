from ultralytics import YOLO

# Load YOLOv8 (nano version, fastest for beginners)
model = YOLO("yolov8n.pt")


```python
results = model("https://ultralytics.com/images/bus.jpg")
results[0].show()
```

```python
results[0].save(filename="bus_output.jpg")
```

```python
# Train YOLO on small dataset (coco128)
model.train(data="coco128.yaml", epochs=3, imgsz=640)
```
